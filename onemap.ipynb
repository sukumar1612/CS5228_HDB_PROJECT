```python
def get_aux_data(query):
    url = "https://www.onemap.gov.sg/api/public/themesvc/retrieveTheme?queryName=" + query
    response = requests.request("GET", url, headers=headers)

    data = json.loads(response.text)
    var = [entry for entry in data["SrchResults"] if "NAME" in entry]

    rows = []
    for s in var:
        name = s.get("NAME")
        latlng = s.get("LatLng", "")

        if latlng and "," in latlng:
            lat, lon = latlng.split(",", 1)
        else:
            lat, lon = None, None

        rows.append({
            "name": name,
            "LATITUDE": lat,
            "LONGITUDE": lon
        })

    df = pd.DataFrame(rows)
    str = "Data/auxiliary-data/" + query + ".csv"
    df.to_csv(str, index=False)
    print(str)
    return None
```

```python
query = ["nationalparks", "moh_hospitals", "communityclubs", "libraries", "childcare", "heritagetrees", "registered_pharmacy"]
```

```python
for q in query:
    get_aux_data(query = q)
```
